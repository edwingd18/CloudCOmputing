Normalizar Datos

```python
def normalizar(x,xmax,xmin,ymax,ymin):
  m = (ymax-ymin)/(xmax-xmin)
  b = ymin - m*xmin

  y = m*x + b

  return y
```

```python
max = data.max().values
min = data.min().values
data2 = normalizar(data,max,min,1,0)
data2.head(10)
```

Creación del modelo MLP


```python

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import plot_model
from sklearn.model_selection import train_test_split
```

```python

target = data2['CO'].values # y_train
data3 = data2.drop(['CO', 'Año', 'Presion_Compresor', 'Presion_Turbina_Gas'], axis=1)


print(target.shape)
print(data3.shape)


```

Separar datos de entrenamiento y validación

```python

seed = 42

x_train,x_test,y_train,y_test = train_test_split(data3, target, test_size=0.2, random_state=seed)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
input_dim = x_train.shape[1]
num_clases = 1
lr = 1e-3
epochs = 800
loss = 'mse' #'mean_squared_error'
```

Creación de la primera arquitectura y optimizador


```python
def crear_modelo_1():
    modelo = Sequential()
    modelo.add(Dense(24, input_dim=input_dim, activation='relu'))
    modelo.add(Dense(16, activation='relu'))
    modelo.add(Dense(num_clases, activation='linear'))

    modelo.summary()

    opt = tf.keras.optimizers.Adam(learning_rate=lr)

    modelo.compile(loss = loss, optimizer = opt)

    return modelo
model = crear_modelo_1()
```

```python
# Asegúrate de que al compilar el modelo uses una función de pérdida válida,
# por ejemplo:
# model.compile(loss='mse', optimizer=opt)

# Reiniciamos las listas de seguimiento sin utilizar el nombre 'loss'
loss_history = []
val_loss_history = []
epochs = 1

# Entrenamiento inicial de 1 época
history = model.fit(x_train, y_train, validation_data=(x_test, y_test),
                    epochs=1, batch_size=32, verbose=0)

loss_history.append(history.history['loss'][0])
val_loss_history.append(history.history['val_loss'][0])

print(f'Epoch: {epochs} - loss: {loss_history[-1]:.6f} - val_loss: {val_loss_history[-1]:.6f}')
epochs += 1

# Entrenar hasta que val_loss sea menor o igual a 0.0010
while val_loss_history[-1] > 0.000640:
    history = model.fit(x_train, y_train, validation_data=(x_test, y_test),
                        epochs=1, batch_size=32, verbose=0)
    loss_history.append(history.history['loss'][0])
    val_loss_history.append(history.history['val_loss'][0])
    print(f'Epoch: {epochs} - loss: {loss_history[-1]:.6f} - val_loss: {val_loss_history[-1]:.6f}')
    epochs += 1


```

### Generar matrices de pesos y bias

```python
import numpy as np

# Función para exportar pesos y bias a C++
def export_weights_to_cpp(model, filename="pesos_y_bias.cpp"):
    weights = model.get_weights()

    cpp_code = "// Pesos y bias generados desde TensorFlow/Keras\n\n"

    for i in range(0, len(weights), 2):  # Iterar sobre capas (pesos y bias vienen en pares)
        layer_idx = i // 2 + 1  # Índice de la capa
        w = weights[i]
        b = weights[i + 1]

        # Formatear pesos
        cpp_code += f"// Pesos de la capa {layer_idx}\n"
        cpp_code += f"float pesos_capa{layer_idx}[{w.shape[0]}][{w.shape[1]}] = {{\n"
        cpp_code += ",\n".join(["{" + ", ".join(f"{val:.6f}" for val in row) + "}" for row in w])
        cpp_code += "\n};\n\n"

        # Formatear bias
        cpp_code += f"// Bias de la capa {layer_idx}\n"
        cpp_code += f"float bias_capa{layer_idx}[{b.shape[0]}] = {{"
        cpp_code += ", ".join(f"{val:.6f}" for val in b)
        cpp_code += "};\n\n"

    # Guardar en archivo
    with open(filename, "w") as f:
        f.write(cpp_code)

    return cpp_code

# Ejecutar la función después de entrenar tu modelo
cpp_code = export_weights_to_cpp(model, "pesos_y_bias.cpp")

# Mostrar código generado
print(cpp_code)
```

### generar vector de entrada

```python
import numpy as np
import pandas as pd

# Función para exportar un vector de entrada a C++
def export_input_vector(x_test, filename_txt="input.txt", filename_cpp="input_vector.cpp"):
    # Seleccionar un ejemplo del conjunto de prueba
    input_vector = x_test.iloc[0].values if isinstance(x_test, pd.DataFrame) else x_test[0]

    # Guardar en un archivo CSV (.txt)
    np.savetxt(filename_txt, input_vector, delimiter=",", fmt="%.6f")

    # Convertir a código C++
    cpp_code = "// Vector de entrada generado desde TensorFlow/Keras\n\n"
    cpp_code += "float input_vector[" + str(len(input_vector)) + "] = {"
    cpp_code += ", ".join(f"{val:.6f}" for val in input_vector)
    cpp_code += "};\n"

    # Guardar en archivo C++
    with open(filename_cpp, "w") as f:
        f.write(cpp_code)

    return cpp_code

# Cargar tus datos de prueba (x_test debe estar definido en tu entorno)
try:
    cpp_code = export_input_vector(x_test, "input.txt", "input_vector.cpp")

    # Mostrar el código C++ generado
    print(cpp_code)

except NameError:
    print("Error: No se encontró x_test en el entorno actual. Asegúrate de que x_test está definido.")
```
